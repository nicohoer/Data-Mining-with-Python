# Clustering with K-Means

The aim of clustering is to find similarities in unlabeled data (unsupervised learning), build groups of similar data and possibly use the result to assign a predicted group to new data.

***

**Short Summary K-Means:**

The K-Means algorithm creates K clusters by minimizing the internal distances within each cluster (called **inertia**). 

## Finding Clusters Of Mall Customers

Dataset created for learning purposes for Udemy's Machine Learning A-Z course.

Data from https://www.kaggle.com/vjchoudhary7/customer-segmentation-tutorial-in-python

```python
import pandas as pd
mall = pd.read_csv("C:/Users/Nicoo/Desktop/Exercise Projects/Data Mining with Python/Mall_Customers.csv")
mall = mall.drop(["CustomerID", "Gender"], axis = "columns")
mall.columns = ["Age", "Income", "Score"]
mall.head(2)
```

**The columns:**

* Age: Age of customer
* Income: Annual Income (in 1,000 USD)
* Score: Score assigned by the mall based on customer behavior and spending nature

## Some Data Exploration:

**Income and Spending Score:**

## Clustering with K-Means:

* the quality of clusters in the K-Means algorithm depends on the number of clusters we have decided to create. A common way to find the appropriate number of clusters is to look for an *elbow* in scree plot:

**First, create multiple models with a different number for K:**

```python
from sklearn.cluster import KMeans
inertia = []
k_list = [2, 3, 4, 5, 6, 7, 8, 9]
for k in k_list:
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(mall)
    inertia.append(kmeans.inertia_)
```

**Then plot the internal distances (inertia) against the number of clusters:**

```python
plt.plot(k_list, inertia)
plt.xlabel("Number of clusters")
plt.ylabel("Internal distances")
plt.show()
```

Both k = 4 and k = 5 seem to be good values, but I've decide to go for k = 5 because of the five  distinct clusters in the Income vs. Score plot.

### Final Model and plotting the Clusters

```python
clustering = KMeans(n_clusters = 5, random_state = 42)
clustering.fit(mall)
clusters = clustering.predict(mall)
```

#### Plotting With Clusters:

**Income and Score:**

```python
plt.scatter(mall["Income"], mall["Score"], c = clusters)
plt.xlabel("Annual Income (in 1,000 USD)")
plt.ylabel("Spending Score (1 - 100)")
plt.show()
```

As predicted theres five  distinct clusters as described in the data exploration.

**Age and Income:**

```python
plt.scatter(mall["Age"], mall["Income"],c = clusters)
plt.xlabel("Age")
plt.ylabel("Income")
plt.show()
```

Despite some outliers there is one cluster for young customers with low income.

**Age and Spending Score:**

```python
plt.scatter(mall["Age"], mall["Score"], c = clusters)
plt.xlabel("Age")
plt.ylabel("Score")
plt.show()
```

There's now two clusters for young customers with high score and a cluster with customers of every age with score between 40 and 60.

**Trying to detect patterns by looking at cluster centers:**

```python
pd.DataFrame(clustering.cluster_centers_, columns = mall.columns)
```
