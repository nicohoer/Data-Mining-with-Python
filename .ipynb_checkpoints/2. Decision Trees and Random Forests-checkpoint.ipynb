# Decision Trees and Random Forests for Regression Problems

In this notebook I will go through some of the theory behind Decision Trees and Random Forests for Regression problems as well as some common steps when modelling them with Python using scikit-learn, pandas and other modules.
<br>
***

**Short Summary: Decision Tree**

<mark>A Decision Tree predicts the outcome for set of data by applying a set of decision rules to the data. Every node represents a condition on the features, which then splits into two branches depending on whether the data fulfills the condition or not. At the end of a branch can either be a new node with a condition or a leaf. A leaf then holds the value of the prediction, which can either be the predicted class for a Classification or a metric value for a Regression. The predicted value for a Regression is the mean of all response values of data points that fall in the range of this leaf.<mark>

<mark>Using Recursive Binary Splitting the Decision Tree tries finding the right feature and threshhold value for that feature which minimizes the cost function of the split. Unrestricted splitting can lead to very deep trees which are overfitted to the data and are bad at generalizing the problem. To prevent overfitting a Decision Tree theres different approaches: setting a maximum depth, setting a minimum amount of data points in every leaf or pruning.<mark>

<mark>One pruning method is called post-pruning which happens after a complete Decision Tree has been modelled. The aim of post-pruning is to remove branches of a big Decision Tree that have low importance or that fulfill other custom criteria, such as accuracy, error scores and others.<mark>

# Predicting House Prices with Regression Trees

## The data

**Getting the data from sklearn.datasets and removing some features**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
columns = ["CRIME", "ZN", "INDUSTY", "CHAS", "NOX", "ROOM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]
X = pd.DataFrame(X, columns = columns)
y = pd.DataFrame(y, columns = ["VALUE"])
X = X.drop(["B", "CHAS", "ZN"], axis = "columns")
X.head(2)
```

### Columns:

* CRIME: per capita crime rate by town
* INDUSTRY: proportion of non-retail business acres per town
* NOX: nitric oxides concentration (parts per 10 million)
* ROOM: average number of rooms per dwelling
* AGE: proportion of owner-occupied units built prior to 1940
* DIS: weighted distances to five Boston employment centres
* RAD: index of accessibility to radial highways
* TAX: full-value property-tax rate per 10,000 USD
* PTRATIO: pupil-teacher ratio by town
* LSTAT: lower status of the population
* **Target: MED_VALUE:  Median value of owner-occupied homes in $1000â€™s**

### Relationship between Features and Response

## 1. Simple Tree with Two Features (max depth 2)

* First tree with only ROOM and LSTAT and max_depth = 2

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
regressor1 = DecisionTreeRegressor(random_state=0, max_depth = 2)
X_reduced = X[["ROOM", "LSTAT"]]
regressor1.fit(X_reduced, y)
fig = plt.figure()
fig.set_figwidth(9)
fig.set_figheight(6)
plot_tree(regressor1, filled=True, feature_names = X_reduced.columns)
plt.show()
```

**Interpretation:**

* Split condition on features (for example ROOM <= 6.941)
* MSE of all data points that are included in this node
* samples: the amount of data points in this node
* value: the predicted value for data that belongs to this node; also the average response of all data in this node

*value of root node equals the average value of the complete dataset:*

```python
print(np.mean(y))
```

*average value for houses with ROOM <= 6.941 and LSTAT <= 14.4 (value for leaf on the most left):*

```python
temp_join = y.join(X_reduced)
np.mean(temp_join.loc[(temp_join["ROOM"] <= 6.941) & (temp_join["LSTAT"] <= 14.4)].VALUE)
```

*there's 175 rows with ROOM <= 6.941 and LSTAT > 14.4:*

```python
(temp_join.loc[(temp_join["ROOM"] <= 6.941) & (temp_join["LSTAT"] > 14.4)]).shape[0]
```

*predicting the House Price for a house with 7 rooms and lstat = 20:*

- ROOM > 6.941 (goes to right child)
- ROOM < 7.347 (goes to left child)
- value at this leaf: 32.113

*Using our model:*

```python
regressor1.predict(np.array([7, 20]).reshape(1, -1))
```

**Evaluating Performance (In-sample):**

R2:

```python
from sklearn.metrics import r2_score
y_pred = regressor1.predict(X_reduced)
r2_score(y, y_pred)
```

MSE:

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_pred)
```

Cross-validation with neg. MSE:

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(regressor1, X_reduced, y, cv=3, scoring = "neg_mean_squared_error"))
```

## Random Forests

**Short Summary Random Forests:**

<mark>Making predictions using only one Decision Tree can be inconsistent and have mediocre results. A better alternative is to use multiple trees and then use the average of all the predictions of the subtrees as the final prediction of the ensemble (<b>Bagging</b>). This can reduce the variance of predictions and make the model more reliable and overall better. To further prevent overfitting the trees it's useful to randomly leave out some features in the subtrees. This way in some trees more important features are left out which forces the model to consider the other features and better generalize the relationship between features and target.
    <br>*This ensemble method of multiple randomized trees is called <b>Random Forest.</b>*<mark>

## 3.Random Forest with All Features

* our final model will be a Random Forest with all features and using pre-pruning parameters; we will properly examine its performance with out-of-sample testing:

- I will use the recommended default parameters max_depth = None and min_samples_split=2
- in every Tree two features will randomly be left out: max_features = X_train.columns.size - 2
- our Forest will consist of 50 trees

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
rfr = RandomForestRegressor(n_estimators = 50, max_features = X_train.columns.size - 2)
rfr.fit(X_train, np.ravel(y_train))
```

*Looking at the feature importances of our final model:*

```python
with plt.style.context("bmh"):
    plt.bar(x = np.arange(10), height = rfr.feature_importances_, width = 0.30)
    plt.xticks(ticks = np.arange(10), labels = X_train.columns, rotation= 55)
    plt.ylabel("Feature importance")
```

**Evaluating Performance(out-of-sample and CV):**

```python
y_pred3 = rfr.predict(X_test)
```

R2:

```python
r2_score(y_test, y_pred3)
```

Single MSE:

```python
mean_squared_error(y_test, y_pred3)
```

Cross Validation neg.MSE:

```python
np.mean(cross_val_score(rfr, X, np.ravel(y), cv=3, scoring = "neg_mean_squared_error"))
```
